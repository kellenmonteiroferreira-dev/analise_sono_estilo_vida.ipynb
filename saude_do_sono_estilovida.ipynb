```python
# 2- média/ moda e mediana  de duração de sono por profissão#
# Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]


```

```python
import pandas as pd
import numpy as np

# Carregar o arquivo CSV
file_path = 'saude_do_sono_estilo_vida.csv'
data = pd.read_csv(file_path)

# Renomear colunas
data = data.rename(columns={
    'ID': 'Identificador',
    'Ocupação': 'Profissão',
    'Pressão sanguíneaaaa': 'Pressão sanguínea',
    'Categoria BMI': 'Categoria IMC'
})

# Função personalizada para calcular a moda
def mode_function(series):
    mode_vals = series.mode()
    if len(mode_vals) == 1:
        return mode_vals[0]
    else:
        return list(mode_vals)

# Calcular média, mediana e moda da duração do sono para cada profissão
sleep_stats = data.groupby('Profissão')['Duração do sono'].agg(['mean', np.median, mode_function])
sleep_stats.plot(kind='bar')

# Renomear a coluna da moda para 'mode' para consistência
sleep_stats.rename(columns={'mode_function': 'mode'}, inplace=True)

# Exibir os resultados
print(sleep_stats)
```

```python
# Exercício 8- Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)


```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv('saude_do_sono_estilo_vida.csv')

# Renomear a coluna de pressão sanguínea
df = df.rename(columns={'Pressão sanguíneaaaa': 'Pressão sanguínea'})

# Separar a pressão sanguínea em duas colunas: sistólica e diastólica
df[['Pressão sistólica', 'Pressão diastólica']] = df['Pressão sanguínea'].str.split('/', expand=True).astype(int)

# Calcular a média da pressão sanguínea para homens
media_pressao_homens = df[df['Gênero'] == 'Homem'][['Pressão sistólica', 'Pressão diastólica']].mean().mean()

# Calcular a média da pressão sanguínea para mulheres
media_pressao_mulheres = df[df['Gênero'] == 'Mulher'][['Pressão sistólica', 'Pressão diastólica']].mean().mean()

# Exibir os resultados
print(f'A média da pressão sanguínea para homens é: {media_pressao_homens}')
print(f'A média da pressão sanguínea para mulheres é: {media_pressao_mulheres}')

# Determinar quem tem maior pressão sanguínea média
if media_pressao_homens > media_pressao_mulheres:
    print('Homens têm maior pressão sanguínea média.')
else:
    print('Mulheres têm maior pressão sanguínea média.')
```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv('saude_do_sono_estilo_vida.csv')

# Renomear as colunas
df = df.rename(columns={'ID': 'Identificador', 'Ocupação': 'Profissão', 'Pressão sanguíneaaaa': 'Pressão sanguínea', 'Categoria BMI': 'Categoria IMC'})

# Calcular a moda da duração do sono
moda_sono = df['Duração do sono'].mode()[0]

# Exibir o resultado
print(f'A moda da duração do sono é: {moda_sono} horas')

# Verificar se a moda é 8 horas
if moda_sono == 8:
    print('É predominante entre os participantes dormir 8 horas por dia.')
else:
    print('Não é predominante entre os participantes dormir 8 horas por dia.')
```

```python
# 10 Exercício - Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média
```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv('saude_do_sono_estilo_vida.csv')

# Renomear as colunas
df = df.rename(columns={'ID': 'Identificador', 'Ocupação': 'Profissão', 'Pressão sanguíneaaaa': 'Pressão sanguínea', 'Categoria BMI': 'Categoria IMC'})

# Filtrar as pessoas com frequência cardíaca acima de 70
acima_70 = df[df['Frequência cardíaca'] > 70]
media_passos_acima_70 = acima_70['Passos diários'].mean()

# Filtrar as pessoas com frequência cardíaca menor ou igual a 70
igual_ou_menor_70 = df[df['Frequência cardíaca'] <= 70]
media_passos_igual_ou_menor_70 = igual_ou_menor_70['Passos diários'].mean()

# Exibir os resultados com 2 casas decimais
print(f'A média de passos diários para pessoas com frequência cardíaca acima de 70 é: {media_passos_acima_70:.2f}')
print(f'A média de passos diários para pessoas com frequência cardíaca menor ou igual a 70 é: {media_passos_igual_ou_menor_70:.2f}')

# Determinar quem dá mais passos em média
resultado = ['Acima de 70', 'Menor ou igual a 70'][media_passos_acima_70 <= media_passos_igual_ou_menor_70]
print(f'Pessoas com frequência cardíaca {resultado} dão mais passos em média.')
```

```python
import pandas as pd

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv('saude_do_sono_estilo_vida.csv')

# Renomear a coluna de pressão sanguínea
df = df.rename(columns={'Pressão sanguíneaaaa': 'Pressão sanguínea'})

# Criar uma nova coluna para categorizar a frequência cardíaca
df['Categoria Frequência'] = df['Frequência cardíaca'].apply(lambda x: 'Acima de 70' if x > 70 else '70 ou Menos')

# Calcular a média de passos diários para cada categoria de frequência cardíaca
media_passos = df.groupby('Categoria Frequência')['Passos diários'].mean()

# Determinar qual categoria tem a maior média de passos diários
categoria_mais_passos = media_passos.idxmax()

# Exibir o resultado
print(f'A categoria com a maior média de passos diários é: {categoria_mais_passos}')
print(media_passos)
```
